```python
import pandas as pd
```

```python
raw_df = pd.read_csv('../../../test-builds/llvm/datasets/model-aug-29.csv')
raw_df.head()
```

```python
cleaned_df = raw_df.copy()

# don't want the 'fullpath', 'modulepath', 'function' columns
cleaned_df.pop(['fullpath', 'modulepath', 'function'])

def clean_labels(x):
    if x == 'O1':
        x = 1 
    elif x == 'O2':
        x = 2 
    elif x == 'O3':
        x = 3
    else:
        x = 4

cleaned_df['label'] = cleaned_df['label'].apply(clean_labels)

cleaned_df.head()        
```
